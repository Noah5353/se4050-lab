```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create 4*4 array from an exponential distribution
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4*4 exponential distribution array:")
print(exp_array)
```

```python
#Visualize exponential, uniform, and normal distribution
exp_data= np.random.exponential(scale=1.0, size=100)
uniform_data= np.random.rand(100000)
normal_data = np.random.randn(100000)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions Comparison")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# For 3D plots
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = X*2 + Y*2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z = X^2 + Y^2")
plt.title("3D Plot: Z = X^2 + Y^2")
plt.show()
```

```python
import pandas as pd
# Pearson and Spearman correlation for [HP, Attack, Defense, Sp. Atk, Sp. Def, Speed]
data= {
    "HP": np.random.randint(50,150,100),  #generates 100 random int between 50 and 150for each stat
    "Attack": np.random.randint(50,150,100),
    "Defense": np.random.randint(50,150,100),
    "Sp. Atk": np.random.randint(50,150,100),
    "Sp. Def": np.random.randint(50,150,100),
    "Speed": np.random.randint(50,150,100),
}
df = pd.DataFrame(data)
```

```python
import seaborn as sns
pearson_corr=df.corr(method="pearson")
spearman_corr= df.corr(method="spearman")

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm")
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm")
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
